```python
import numpy as np
import math
from matplotlib import pylab as plt
```

```python
x = np.arange(-10, 10, 0.1)
y = x ** 3
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend('G')
plt.show()
```

```python
from matplotlib import rc
rc('text', usetex=True)
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
linear = plt.plot(x,x)
square = plt.plot(x, x ** 2)
cos = plt.plot(x, [math.cos(elem) for elem in x])
sin = plt.plot(x, [math.sin(elem) for elem in x])
nabs = plt.plot(x, [abs(elem) for elem in x])
plt.legend([r'$y=x$', r'$y=x^2$', r'$y=cos(x)$', r'$y = sin(x)$', r'$y = \abs{x}$ '])
plt.show()
```
